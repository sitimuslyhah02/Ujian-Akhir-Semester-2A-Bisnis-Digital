```python
# ==============================
# Toko Baju Online GOFLI V2.0
# ==============================

produk = {
    "Kaos Polos": {"harga": 50000, "stok": 10},
    "Kemeja Flanel": {"harga": 120000, "stok": 5},
    "Jaket Denim": {"harga": 250000, "stok": 3},
    "Celana Jeans": {"harga": 180000, "stok": 8}
}

keranjang = {}

def tampilkan_menu():
    print("\n===== Toko Baju Online GOFLI =====")
    print("1. Lihat Daftar Produk")
    print("2. Cari Produk")
    print("3. Tambah ke Keranjang")
    print("4. Lihat Keranjang")
    print("5. Hapus dari Keranjang")
    print("6. Checkout")
    print("7. Keluar")

def tampilkan_produk():
    print("\n--- Daftar Produk ---")
    for nama, detail in produk.items():
        print(f"- {nama}: Rp{detail['harga']} (Stok: {detail['stok']})")
    print("----------------------")

def cari_produk():
    keyword = input("Masukkan nama/kata kunci produk: ").lower()
    ditemukan = False
    print("\n--- Hasil Pencarian ---")
    for nama, detail in produk.items():
        if keyword in nama.lower():
            print(f"- {nama}: Rp{detail['harga']} (Stok: {detail['stok']})")
            ditemukan = True
    if not ditemukan:
        print("Produk tidak ditemukan.")
    print("------------------------")

def tambah_ke_keranjang():
    tampilkan_produk()
    nama = input("Masukkan nama produk yang ingin dibeli: ")
    if nama in produk:
        try:
            jumlah = int(input("Masukkan jumlah: "))
            if jumlah > 0 and jumlah <= produk[nama]["stok"]:
                keranjang[nama] = keranjang.get(nama, 0) + jumlah
                produk[nama]["stok"] -= jumlah
                print(f"‚úÖ {jumlah} {nama} berhasil ditambahkan ke keranjang.")
            else:
                print("‚ùå Jumlah tidak valid atau stok tidak cukup.")
        except ValueError:
            print("‚ùå Masukkan angka yang valid.")
    else:
        print("‚ùå Produk tidak ditemukan.")

def lihat_keranjang():
    print("\n--- Keranjang Belanja ---")
    if not keranjang:
        print("Keranjang Anda masih kosong.")
    else:
        total = 0
        for nama, jumlah in keranjang.items():
            harga = produk[nama]["harga"]
            subtotal = harga * jumlah
            total += subtotal
            print(f"- {nama} ({jumlah}x): Rp{subtotal}")
        print("--------------------------")
        if total > 500000:
            diskon = int(total * 0.10)
            total -= diskon
            print("üéÅ Diskon 10%: -Rp{}".format(diskon))
        print(f"Total Belanja: Rp{total}")
    print("--------------------------")

def hapus_dari_keranjang():
    lihat_keranjang()
    nama = input("Masukkan nama produk yang ingin dihapus: ")
    if nama in keranjang:
        try:
            jumlah = int(input("Masukkan jumlah yang ingin dihapus: "))
            if jumlah >= keranjang[nama]:
                produk[nama]['stok'] += keranjang[nama]
                del keranjang[nama]
                print(f"{nama} dihapus dari keranjang.")
            else:
                keranjang[nama] -= jumlah
                produk[nama]['stok'] += jumlah
                print(f"{jumlah} {nama} dihapus dari keranjang.")
        except ValueError:
            print("‚ùå Masukkan angka yang valid.")
    else:
        print("‚ùå Produk tidak ada di keranjang.")

def checkout():
    if not keranjang:
        print("‚ùå Keranjang kosong. Tidak bisa checkout.")
        return
    lihat_keranjang()
    konfirmasi = input("Lanjut ke pembayaran? (ya/tidak): ").lower()
    if konfirmasi == "ya":
        print("üéâ Pembayaran berhasil! Terima kasih telah berbelanja.")
        keranjang.clear()
    else:
        print("Checkout dibatalkan.")

def main():
    while True:
        tampilkan_menu()
        pilihan = input("Masukkan pilihan (1-7): ")
        if pilihan == '1':
            tampilkan_produk()
        elif pilihan == '2':
            cari_produk()
        elif pilihan == '3':
            tambah_ke_keranjang()
        elif pilihan == '4':
            lihat_keranjang()
        elif pilihan == '5':
            hapus_dari_keranjang()
        elif pilihan == '6':
            checkout()
        elif pilihan == '7':
            print("üëã Terima kasih! Sampai jumpa.")
            break
        else:
            print("‚ùå Pilihan tidak valid. Coba lagi.")

if __name__ == "__main__":
    main()
```
